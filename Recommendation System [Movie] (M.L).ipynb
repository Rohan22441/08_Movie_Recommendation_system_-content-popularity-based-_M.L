# Content Based Model :

```python
d1='No Related content found'
try:
    name=input('Enter the name of the movie : ')
    idx=ind[name]
    sig_sc=list(enumerate(sig[idx]))
    sig_sc=sorted(sig_sc,key=lambda w:w[1],reverse=True)
    sig_sc=sig_sc[1:11]
    mo=[m[0] for m in sig_sc]
    d=df5[['original_title','vote_average','vote_count']].iloc[mo]
    d.rename(columns={'original_title':'movies','vote_average':'ratings',
                      'vote_count':'total_votes'},inplace=True)
except:
    print('Movie not found')
else:    
    print('Top 10 Related Contents')
    d1=pd.DataFrame(d.values,columns=d.columns)
'''Number of related Content'''    
d1
```

# Popularity Based Model :

```python
R=df5['vote_average']
v=df5['vote_count']
C=np.mean(df5['vote_average'])
m=df5['vote_count'].quantile(0.7)
```

```python
df5['weighted_avg']=((R*v)+(C*m))/(v+m)
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
f_data=pd.DataFrame(minmax.fit_transform(df5[['weighted_avg','popularity']]))
f_data.columns=['weighted_avg','popularity']
f_data.insert(0,'movies',df5['original_title'].values)
f_data['score']=f_data['weighted_avg']*0.5 + f_data['popularity']*0.5
f_data.sort_values('score',ascending=False,inplace=True)
data=f_data[['movies','score']].head(10)
data
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data['score'],data['movies'])
plt.xlabel('score')
plt.ylabel('movies')
plt.show()
```
